```python
from datetime import datetime
for i in range(50):
    raw_lc_df['last_payment_date'][i] = datetime.strptime(raw_lc_df['last_payment_date'][i], '%b/%d/%Y')
    raw_lc_df['issue_date'][i] = datetime.strptime(raw_lc_df['issue_date'][i], '%b/%d/%Y')
    print ("new datetime row")
```

```python
raw_lc_df['y_stanford'] = (raw_lc_df['total_pymnt']/raw_lc_df['loan_amnt'])**(1/(365/))
```
